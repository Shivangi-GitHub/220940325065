## Decorator

```python
# Decorator

def decor(func):
  def inner(name):
    if name=="Ram":
      print("Hello Ram Good Morning")
    else:
      func(name)

  return inner

@decor
def wish(name):
  print("Hii", name, "Good Morning")

wish("Ram")
wish("Sita")
```

### Decorator Chanining

```python
def decor1(func):
  def inner(name):
    if name=="Lakhan":
      print("Hello Lakhan Good Morning")
    else:
      func(name)

  return inner

def decor2(func):
  def inner(name):
    if name=="Ram":
      print("Hey Ram Good Morning")
    else:
      func(name)

  return inner

@decor2  #----> first this decorator is executed
@decor1  #-----> second time this decorator will get executed
def wish(name):
  print("Hii", name, "Good Morning")

wish("Lakhan")
wish("Ram")
wish("Ram")
wish("Sita")
wish("Shyam")
```

```python
# fibonnacci series using generator

def fib():
  a,b=0,1
  while True:
    yield a
    a,b=b,a+b

f=fib()
print(next(f))
print(next(f))
print(next(f))
print(next(f))
print(next(f))

```

### creating iter and next in class

```python
class MyIterable:
    def __init__(self, data):
        self.data = data
        
    def __iter__(self):
        self.index = 0
        return self
        
    def __next__(self):
        if self.index >= len(self.data):
            raise StopIteration
        value = self.data[self.index]
        self.index += 1
        return value
```

```python
class car():
      
    # init method or constructor
    def __init__(self, model, color):
        print(id(self))
        self.model = model
        self.color = color
          
    def show(self):
        print(id(self))
        print("Model is", self.model )
        print("color is", self.color )
          
# both objects have different self which 
# contain their attributes
audi = car("audi a4", "blue")
ferrari = car("ferrari 488", "green")

print(id(audi))
print(id(ferrari))
  
audi.show()     # same output as car.show(audi)
ferrari.show() 

```

## Destructor

```python
class Employee:
  
    # Initializing
    def __init__(self):
        self.a="A"
        print('Employee created')
  
    # Calling destructor
    def __del__(self):
        print("Destructor called")
  
def Create_obj():
    print('Making Object...')
    obj = Employee()
    print('function end...')
    return obj
  
print('Calling Create_obj() function...')
obj = Create_obj()
print('Program End...')
del obj
```

```python
rows = int(input("Enter the number of rows: "))

for i in range(0, rows):
    for j in range(rows, i,-1):
        print(" ", end="")
    for k in range(0, 2*i+1):
        print("*", end="")
    print()

for i in range(rows-1, 0, -1):
    for j in range(0, rows-i+1):
        print(" ", end="")
    for k in range(1, 2*i):
        print("*", end="")
    print()
```
